Make it print "Even" for even numbers

"Odd" for odd numbers

`````python
def even_odd(x):
    if x % 2 == 0:
        return 'Even'
    # If the function hasn't returned yet, the number is odd
    return 'Odd'
for i in range(1,11):
    result = even_odd(i)
    print(result)
`````
