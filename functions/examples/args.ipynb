```python
def greet(names):
    for name in names:
        print('Hello ', name)
# the argument is in the form of a tuple, so the argument is one.
# inside tuple there are two names but they are dealt as one argument 
# because they are inside one tuple
greet(('Ali', 'Meena'))
```

```python
def greet(*names):
    for name in names:
        print('Hello ', name)
        # here are two positional arguments. not one tuple, so both are dealt separately 
        # due to the presence of asterik.
        # if here were no asterik, it would give a positional argument error.
greet('Ali', 'Meena')

        
```

- The loop is necessary because it helps to iterate over the arguments.

- In case of no asterik, the names(inside tuple )were printed due to iteration over them       inside the tuple and the tuple was dealth as ONE positional argument. That's why it worked.

- Asterik controls how positional arguments are collected at call time. In case of *args, python makes the tuple on its own.

- They are dealt independently. Loops print them separately.


```python
def greet(*names):
    print(*names)
    for name in names:
        print('Hello ', name)
greet('Ali', 'Meena')
greet('Ali', 'Meena', 'Laiba')
```
